```python
def dotproduct(xs, ys):
    total = 0
    for i in xrange(min(len(xs),len(ys))):
        total += xs[i]*ys[i]
    return total       

from random import randrange
xs = [randrange(0,1000) for _ in xrange(1000)]
ys = [randrange(0,1000) for _ in xrange(1000)]

% timeit dotproduct(xs, ys)
```

```python
from numba import autojit

dotproduct_numba = autojit(dotproduct)
% timeit dotproduct_numba(xs, ys)
```

```python
from numpy import ndarray

def mersenne(N):
    seed, period, length = 1, 397, 624
    #state = ndarray(shape=(length,), dtype=int)
    #state[0] = seed & 0xffffffff
    state = [seed & 0xffffffff]
    for i in range(1,length):
        #state[i] = (0x6c078965 * (state[i-1] ^ (state[i-1] >> 30)) + i) & 0xffffffff
        state.append((0x6c078965 * (state[-1] ^ (state[-1] >> 30)) + i) & 0xffffffff)
   
    #rv = ndarray(shape=(N,), dtype=int)
    rv = []
    n = 0    
    while True:
        for i in range(length):
            y = (state[i] & 0x80000000) + (state[(i+1)%length] & 0x7fffffff)
            state[i] = state[(i+period)%length] ^ (y >> 1)
            if y % 2:
                state[i] ^= 0x9908b0df
                
        for i in range(length):            
            if n+1 > N:
                return rv
        
            y = state[i]
            y ^= y >> 11
            y ^= (y << 7) & 0x9d2c5680
            y ^= (y << 15) & 0xefc60000
            y ^= y >> 18
            #rv[n] = y
            rv.append(y)
            n += 1

print mersenne(3)
%timeit mersenne(1000)
```

```python
from numba import autojit

class quadratic(object):
    'simple quadratic (polynomial)'
    def __init__(self, a, b, c):
        self.a, self.b, self.c = a, b, c
    def __add__(self, other):
        return type(self)(self.a + other.a, self.b + other.b, self.c + other.c)

def f(x, y):
    return x + y
g = autojit(f)

# (x+1)(x+1) = x^2 + 2x + 1
# (x+1)(x-1) = x^2 - 1
x, y = quadratic(1, 2, 1), quadratic(1,0,1)

% timeit f(x,y)
% timeit g(x,y)
```
